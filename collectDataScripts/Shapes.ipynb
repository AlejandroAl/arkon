```python
def getListGeoShapes(array):
    keyList="coordinates"
    dictx = eval(array)
    
    return dictx[keyList][0]
```

```python
def generateShapesByTownHall(df, nameColumnTown, nameColumnGeoShape):
    shapesList = df[nameColumnGeoShape].map(getListGeoShapes)
    nameTownList = df[nameColumnTown].values
    
    listTuple = []
    
    for x,y in enumerate(shapesList):
        listTuple.append((nameTownList[x],Polygon(y)))
        
    return listTuple
    
    
```

```python
listShapesTownHall = generateShapesByTownHall(df,"NOMBRE","Geo Shape")
```

```python
pickle_out = open("shapesTownHall.pickle","wb")
pickle.dump(listShapesTownHall, pickle_out)
pickle_out.close()
```

```python
def getTownHall(lat,long):
    pickle_in = open("shapesTownHall.pickle","rb")
    listTuple = pickle.load(pickle_in)
    townHall = ""
    n = len(listTuple)-1
    for nom,polygon in listTuple:
        point = Point(long,lat)
        if polygon.contains(point):
            townHall = nom

    return townHall
```
